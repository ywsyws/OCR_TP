```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.svm import LinearSVC, SVC
from sklearn.multiclass import OneVsRestClassifier
from sklearn.multiclass import OneVsOneClassifier
```

```python
df = pd.read_csv("data_temp/Dataset_feuilles_1.csv")
df_test = pd.read_csv("data_temp/test.csv")
print(df.shape)
print(df_test.shape)
```

```python
df.head()
```

```python
df_test.head()
```

```python
labelencoder=LabelEncoder()
df['species_code'] = labelencoder.fit_transform(df.species)

df.head()
```

```python
X = df.iloc[:,2:-1]
y = df.iloc[:,-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42
)
print(len(X_train), len(X_test), len(y_train), len(y_test))
```

# Baseline model: KNN

```python
params = {'n_neighbors': [1, 2, 3, 4] }
knn = KNeighborsClassifier()

knn_gs = GridSearchCV(knn, params, cv=4)
knn_gs.fit(X_train, y_train)

print(knn_gs.best_params_)
```

```python
# On récupère la prédiction de la valeur positive
knn = KNeighborsClassifier(1)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

# On créé un vecteur de prédiction à partir du vecteur de probabilités
knn_acc = accuracy_score(y_test, y_pred)

print(knn_acc)
```

# OVO

```python
# clf = OneVsRestClassifier(SVC()).fit(X, y)
ovo = OneVsOneClassifier(SVC())

params = {
    "estimator__C": np.logspace(-3, 3, 7),
    "estimator__kernel": ["poly", "rbf"],
    "estimator__degree": [1, 2, 3, 4],
}

gs_ovo = GridSearchCV(ovr, params, cv=3)
gs_ovo.fit(X_train, y_train)

print(gs_ovo.best_params_)
```

```python
# On récupère la prédiction de la valeur positive
ovo = OneVsOneClassifier(SVC(C=10, degree=2, kernel='poly'))
ovo.fit(X_train, y_train)
y_pred = ovo.predict(X_test)

# On créé un vecteur de prédiction à partir du vecteur de probabilités
ovo_acc = accuracy_score(y_test, y_pred)

print(ovo_acc)
```

# OVR

```python
# clf = OneVsRestClassifier(SVC()).fit(X, y)
ovr = OneVsRestClassifier(SVC())

params = {
    "estimator__C": np.logspace(-3, 3, 7),
    "estimator__kernel": ["poly", "rbf"],
    "estimator__degree": [1, 2, 3, 4],
}

gs_ovr = GridSearchCV(ovr, params, cv=3)
gs_ovr.fit(X_train, y_train)

print(gs_ovr.best_params_)
```

```python
# On récupère la prédiction de la valeur positive
ovr = OneVsRestClassifier(SVC(C=10, degree=2, kernel='poly'))
ovr.fit(X_train, y_train)
y_pred = ovr.predict(X_test)

# On créé un vecteur de prédiction à partir du vecteur de probabilités
ovr_acc = accuracy_score(y_test, y_pred)

print(ovr_acc)
```

# SVC

```python
svm = LinearSVC(multi_class="crammer_singer")

params = { 'C': np.logspace(-3, 3, 7), 'penalty':['l1','l2']}

gs_svm = GridSearchCV(svm, params, cv=3)
gs_svm.fit(X_train, y_train)

print(gs_svm.best_params_)
```

```python
# On récupère la prédiction de la valeur positive
svm = LinearSVC(multi_class="crammer_singer", C=1000, penalty='l1')
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test)

# On créé un vecteur de prédiction à partir du vecteur de probabilités
svm_acc = accuracy_score(y_test, y_pred)

print(svm_acc)
```
